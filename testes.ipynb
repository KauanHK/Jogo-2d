```python
import pandas as pd
import random
import seaborn as sns

y = 500
limite = 80
repeticoes = 20

tabela = {'vezes': [i for i in range(repeticoes)],
          'y': []}

for _ in range(repeticoes):

    opcoes = [i for i in range(y - limite, y + limite+1)]
    y = random.choice(opcoes)

    y = 20 if y < 20 else y
    y = 800 if y > 800 else y

    tabela['y'].append(y)

tabela = pd.DataFrame(tabela)

def grafico_line_plot(tab,x,y, titulo: str | None=None,labelx: str | None=None, labely: str | None=None):
    ax = sns.lineplot(data=tab, x=x, y=y)
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax.figure.set_size_inches(12, 6)
    ax.set_title(titulo, loc = 'left', fontsize = 18)
    if labelx:
        ax.set_xlabel(labelx, fontsize = 14)
    if labely:
        ax.set_ylabel(labely, fontsize = 14)
    ax = ax

grafico_line_plot(tabela,x='vezes',y='y')
```
